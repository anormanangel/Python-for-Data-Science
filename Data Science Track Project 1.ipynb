```python
# Finding correlation within dataset qnd visualizing using the dataset
# Compute pairwise correlation of columns, excluding NA/null values.

corr=df.corr()
sns.heatmap(corr, annot = True)
```

```python
# Box plot for dataset showing how amount of alhol affects the whine quality

# A boxplot is a standardized way of displaying the distribution of data based on a five number summary 
# (minimum, first quartile (Q1), median, third quartile (Q3), and maximum). 

# Box plots divide the data into sections that each contain approximately 25% of the data in that set. 
# Box plots are useful as they provide a visual summary of the data enabling researchers to quickly identify mean values

#The more the alcohol Content, the better the wine Quality


sns.boxplot(y=df['alcohol'], x=df['quality'])
```
